##**HW2 คำนวณเวลาเป็นวินาทีของเวลาต่อไปนี้โดยใช้คำสั่ง split() ช่วย (print ออกมาให้สวยงาม)**
- 12:30:15
- 13:41:07
- 09:11:50
- 21:30:15
- 03:10:15

```python
#ตัวอย่าง 01:02:03 

a = (1*60*60) + (2*60) + 3
print(f'01:02:03 แปลงเป็นวินาที ได้ {a} วินาที')
```

```python
## ข้อ1 12:30:15

H1 = (12*60*60) + (30*60) +15
print(f'12:30:15 แปลงเป็นวินาที ได้ {H1} วินาที')
```

```python
## ข้อ2 13:41:07

H2 = (13*60*60) + (41*60) +7
print(f'13:41:07 แปลงเป็นวินาที ได้ {H2} วินาที')
```

```python
## ข้อ3 09:11:50

H3 = (9*60*60) + (11*60) +50
print(f'09:11:50 แปลงเป็นวินาที ได้ {H3} วินาที')
```

```python
## ข้อ4 21:30:15

H4 = (21*60*60) + (30*60) +15
print(f'21:30:15 แปลงเป็นวินาที  ได้ {H4} วินาที')
```

```python
## ข้อ5 03:10:15

H5 = (3*60*60) + (10*60) +15
print(f'03:10:15 แปลงเป็นวินาที ได้ {H5} วินาที')
```
