##**ตารางที่ 1**
>date เป็น YYYY-MM-DD

```python
data1 = os.path.join(path,'alldb.csv') 
data1 = pd.read_csv(data1,parse_dates=['date']) #date เป็น YYYY-MM-DD
data1
```

##**ตวจสอบขนาด**

```python
data1.shape
```

เนื่องจากข้อมูลมีจำนวนมาก แต่เราไม่ได้ใช้ทั้งหมด เพื่อง่ายต่อการทำ prepocessing ดังนั้นจะตัดมาแค่ข้อมูลที่ต้องการใช้จริงๆ ตามโจทย์


```python
data2 = data1[['user','RowKey','date','phone','site','time','urine','title','ov','PartitionKey','age']] 
data2
```

##**ตวจสอบว่ามี NaN หรือ Missing หรือไม่**

```python
data2.isnull().any() 
#.isnull() เพื่อระบุว่านิพจน์มีค่า Null หรือไม่
#.any() เพื่อดูรายละเอียดการเปรียบเทียบข้อมูลให้ออกมาในรูป true false
```

จากผลเราจะพิจารณาจากการอ่านค่า True และ False ถ้า True คือ มีค่า missing ดังนั้น จะเห็นได้ว่าข้อมูลมีค่า missing เกือบในทุกคอลัมน์ ดังนั้น สามารถอนุมานได้เลยว่า ข้อมูลมีค่า missing จำนวนมาก จึงไม่เหมาะกับตัดข้อมูล หรือ .dropna
จึงพิจารณาให้ใช้วิธีแทนค่าเฉพาะคอลัมน์ข้อมูลที่เราต้องการใช้ 

##**ลองใช้ .dropna ซึ่งจะใช้ตัดแถวข้อมูลที่มีค่า missing**

```python
data_drop = data2.dropna() #ใช้ dropna เพื่อต้องการลบข้อมูลทั้งแถวออกเมื่อมีคอลัมน์ใดคอลัมน์หนึ่งเป็น NaN
data_drop
```

```python
removes = data2.shape[0] - data_drop.shape[0]
print(f'size before drop = {data2.shape[0]}')
print(f'size after drop = {data_drop.shape[0]}')
print(f'% from drop = {(removes/data2.shape[0])*100}')
```

ดังนั้น จากการ dropna จะเห็นได้ว่าค่าข้อมูลหายไปเพียงแค่ 4.75 เปอร์เซ็นเท่านั้น ซึ่งอยู่ในช่วงที่ยอมรับได้ จึงสามารถใช้ค่าข้อมูลจากการ dropna ได้ 

>เปลี่ยนชนิดข้อมูล

```python
import numpy as np
```

เปลี่ยนค่าที่เราต้องการเป็น int หรือเลขจำนวนเต็มก่อน โดยกำหนด error='coerce' จะแปลงค่าที่ไม่ใช่ตัวเลขเป็น NaN

```python
data_drop.dtypes
```

```python
data_drop['user'] = pd.to_numeric(data_drop['user'], errors='coerce')
data_drop['site'] = pd.to_numeric(data_drop['site'], errors='coerce')
data_drop['ov'] = pd.to_numeric(data_drop['ov'], errors='coerce')
data_drop['PartitionKey'] = pd.to_numeric(data_drop['PartitionKey'], errors='coerce')
data_drop['age'] = pd.to_numeric(data_drop['age'], errors='coerce')
```
